```python
import numpy as np
```

1. Примеры создания массивов

Из обычных списков

```python
arr_list = np.array([1, 2, 3, 4, 5])
arr_list
```

Использование zeros, ones, arange

```python
arr_zeros = np.zeros(5)
arr_ones = np.ones(5)
arr_arange = np.arange(0, 10, 2)

arr_zeros, arr_ones, arr_arange
```

Демонстрация атрибутов shape и dtype

```python
print("Форма массива:", arr_arange.shape)
print("Тип данных:", arr_arange.dtype)
```

2.Индексация и срезы

Получение отдельного элемента

```python
element = arr_list[2]   # элемент с индексом 2
element
```

Извлечение подмассива с помощью среза

```python
sub_array = arr_list[1:4]  # элементы со 2-го по 4-й
sub_array
```

3. Векторные операции и базовая статистика

Сложение двух массивов

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
a_plus_b = a + b
a_plus_b
```

Умножение двух массивов

```python
a_times_b = a * b
a_times_b
```

Среднее значение

```python
mean_value = np.mean(a)
mean_value
```
